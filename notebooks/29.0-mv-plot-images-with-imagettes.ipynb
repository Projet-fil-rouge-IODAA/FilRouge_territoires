```python
# get image from data/raw

import rasterio as rio
import numpy as np
import matplotlib.pyplot as plt

path_image = "../data/raw/crop_SENTINEL2A_20151226-111142-750_L2A_T31UDQ_D_V1-1.tif"

src = rio.open(path_image)
red = src.read(2)
green = src.read(3)
blue = src.read(4)
nri = src.read(1)

redn = (red/6).astype(int)
greenn = (green/6).astype(int)
bluen = (blue/6).astype(int)
nirn = (nri/6).astype(int)


# Create RGB natural color composite
rgb = np.dstack((redn, greenn, bluen))

# Let's see how our color composite looks like
plt.figure(figsize=(10,10))

plt.imshow(rgb)

# add a rectangle on the given coordinates

# Create a Rectangle patch
coords = (534, 480, 584, 500)
rect = plt.Rectangle((coords[0], coords[1]), coords[2]-coords[0], coords[3]-coords[1], linewidth=1, edgecolor='r', facecolor='none')
plt.gca().add_patch(rect)

coords = (422, 399, 472, 419)
rect = plt.Rectangle((coords[0], coords[1]), coords[2]-coords[0], coords[3]-coords[1], linewidth=1, edgecolor='r', facecolor='none')
plt.gca().add_patch(rect)

coords = (255, 82, 305, 102)
rect = plt.Rectangle((coords[0], coords[1]), coords[2]-coords[0], coords[3]-coords[1], linewidth=1, edgecolor='r', facecolor='none')
plt.gca().add_patch(rect)

plt.show()
```
